```python
## 전처리
def drop_str(data):
    cnt=0
    for y,m,d,t in zip(data["Year"],data["Month"],data["Date"],data["Time"]):
            data["Year"][cnt] = y.split("년")[0]
            data["Month"][cnt] = m.split("월")[0]
            data["Date"][cnt] = d.split("일")[0]
            data["Time"][cnt] = t.split("시")[0]
            cnt+=1
    return data
def str2int(data):
    data["Year"]=0
    data["Month"]=0
    data["Date"]=0
    data["Time"]=0
    for i, element in enumerate(data["일시"]):
        data["Year"][i]=element.split(' ')[0]
        data["Month"][i]=element.split(' ')[1]
        data["Date"][i]=element.split(' ')[2]
        data["Time"][i]=element.split(' ')[3]
    return data
def cat_1(data):
    strength_map={}
    for i, s in enumerate(data["강도"].unique()):
        strength_map[s] = i
    data["강도"] = data["강도"].map(strength_map)
    return data
def cat_2(data):
    for i, s in enumerate(data["크기"]):
        if s=="-":
            data["크기"][i] = 0
        elif s=="소형":
            data["크기"][i] = 1
        elif s=="중형":
            data["크기"][i] = 2
        else:
            data["크기"][i] = 3
    return data
def prepro(data):
    idx = data[data["Unnamed: 1"]!=0].index
    data = data.drop(idx)
    data = data.reset_index()
    data = str2int(data)
    drop_str(data)
    data =cat_1(data)
    data =cat_2(data)
    data = data.drop(["진행방향","70% 확률 반경(km)", "Unnamed: 1","일시","index"], axis=1)
    return data
```

### LSTM Encoder


```python
# class encode_lstm(nn.Module):
#     def __init__(self, input_size, hidden_size, num_layer):
#         super(encode_lstm, self).__init__()
#         self.input_size = input_size
#         self.hidden_size = hidden_size
```

```python
class simpleGRU(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layer=1):
        super(simpleGRU, self).__init__()
        self.input_size = input_size
        # self.embedding_size = embedding_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.num_layer = num_layer
        
        # self.encoder = nn.Embedding(input_size, embedding_size)
        self.rnn = nn.GRU(input_size, hidden_size, num_layer, batch_first=True)
        self.decoder = nn.Linear(hidden_size, output_size)
        
    def forward(self, x, hidden):
        # embed = self.encoder(x)
        out, hidden = self.rnn(x, hidden)
        out = self.decoder(out)
        
        return out, hidden

    def init_hidden(self, batch_size):
        hidden = torch.randn(self.num_layer, batch_size, self.hidden_size).to(device)
        return hidden
    
    
model = simpleGRU(input_size, hidden_size, output_size, num_layer)
model.to(device)
```

```python
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)
```

```python
epoch = 200

valid_loss_min = np.Inf
# Save training loss
train_loss = torch.zeros(epoch)
# Save validation loss
valid_loss = torch.zeros(epoch)

for e in trange(epoch):
    # Training Session
    model.train()
    for data, target in train_loader:
        # Initialize hidden state
        hidden = model.init_hidden(data.shape[0])
        # Data to GPU
        data, target = data.to(device), target.to(device)
        # Initialize gradients
        optimizer.zero_grad()
        # Calculate logits & hidden
        logits, hidden = model(data, hidden)
        # Calculate loss
        loss = criterion(logits, target)
        # Backpropagation
        loss.backward(retain_graph=True)
        # Update gradient
        optimizer.step()
        # Calculate loss per batch
        train_loss[e] = train_loss[e] + loss.item()
    # Calculate loss per epoch
    train_loss[e] = train_loss[e] / len(train_loader)
    
    # Validation Session
    with torch.no_grad():
        model.eval()
        for data, target in valid_loader:
            # Initialize hidden state
            hidden = model.init_hidden(data.shape[0])
            # Data to GPU
            data, target = data.to(device), target.to(device)
            # Calculate logits & hidden
            logits, hidden = model(data, hidden)
            # Calculate loss (MSE)
            loss = criterion(logits, target)
            # Calculate loss per batch
            valid_loss[e] =  valid_loss[e] + loss.item()
        # Calculate loss per epoch
        valid_loss[e] =valid_loss[e] / len(valid_loader)
    
    print("Epoch : {}\tTraining Loss: {:.6f}\tValidation Loss: {:.6f}".format(e, train_loss[e], valid_loss[e]))
    
    if valid_loss[e] <= valid_loss_min:
        print('Validation loss decreased ({:.6f} ==> {:.6f}).\tSaving model ...'.format(valid_loss_min, valid_loss[e]))
        valid_loss_min = valid_loss[e]
```
